Tên đề tài chung đã được xác định là **"Các Mô hình Học sâu Lai và Tích hợp Trí tuệ Nhân tạo (AI) sử dụng Dữ liệu Đa phương thức và Địa không gian trong Dự đoán và Quản lý Rủi ro Thảm họa"**.

Dưới đây là tổng quan tình hình nghiên cứu liên quan đến đề tài này, dựa trên các nguồn được cung cấp:

### Tình hình Nghiên cứu Liên quan đến Đề tài

Nghiên cứu về dự đoán và quản lý thảm họa đang chuyển dịch mạnh mẽ từ các phương pháp truyền thống sang việc tích hợp Trí tuệ Nhân tạo (AI), Học máy (ML), Học sâu (DL), và Phân tích Địa không gian (Geospatial Analysis) trên phạm vi toàn cầu.

#### I. Tổng quan tình hình nghiên cứu trong và ngoài nước

**1. Các Phương pháp Truyền thống và Hạn chế**

Các phương pháp dự báo thảm họa truyền thống chủ yếu dựa vào **phân tích chuỗi thời gian** (time series analysis), **mô hình hồi quy** (regression models), **mô hình thủy văn** (hydrological models), **hệ thống chuyên gia** (expert systems), và **phân tích thống kê** dựa trên dữ liệu lịch sử,,.
*   **Hạn chế:** Các phương pháp này thường bị giới hạn bởi:
    *   Khả năng xử lý **các mối quan hệ phi tuyến tính và hỗn loạn** (chaotic, non-linear relationships) của thảm họa.
    *   Khả năng **kết hợp dữ liệu thời gian thực** (real-time data) và **thích ứng với các điều kiện thay đổi nhanh chóng**,,.
    *   Yêu cầu dữ liệu lớn, tính toán phức tạp, và khó khăn trong việc xác định các tham số mô hình đối với thiên tai như lũ quét.
    *   Dễ bị ảnh hưởng bởi **sai lệch (bias)** và có **khả năng mở rộng hạn chế** (limited scalability).

**2. Sự trỗi dậy của AI/Học máy trong Quản lý Thảm họa (Xu hướng Hiện đại)**

Việc sử dụng các công nghệ AI/Học máy và Học sâu đã tạo ra những cơ hội mới để cải thiện quản lý thảm họa, cho phép **dự đoán chính xác hơn**, **giám sát tốt hơn**, và **lập kế hoạch chiến lược hiệu quả hơn**,.
*   **Học máy truyền thống:** Các thuật toán như Random Forest (RF), Support Vector Machines (SVM), và Gradient Boosting (GB) đã được chứng minh là hiệu quả trong việc **phân tích các tập dữ liệu lớn** để nhận dạng xu hướng và dự đoán thảm họa, đặc biệt RF rất hiệu quả khi xử lý các tập dữ liệu lớn và phức tạp,. SVM cũng xuất sắc trong các nhiệm vụ phân loại.
*   **Học sâu (Deep Learning):** Các kiến trúc như **Mạng nơ-ron tích chập (CNNs)** và **Mạng nơ-ron hồi tiếp (RNNs/LSTMs)** rất hiệu quả trong việc xử lý dữ liệu cấu trúc, hình ảnh, hoặc chuỗi thời gian,,,.
    *   CNNs giúp **phân tích ảnh vệ tinh** để đánh giá thiệt hại theo thời gian thực,,.
    *   RNNs (ví dụ: LSTM, GRU) được sử dụng để **phân tích dữ liệu tuần tự** như diễn biến mực nước sông hoặc cường độ bão nhiệt đới theo thời gian,,,.
*   **Tích hợp Dữ liệu Địa không gian (GeoAI):** Sự kết hợp giữa AI và Phân tích Địa không gian (Geospatial Analysis) đã được chứng minh là vô giá trong quản lý thảm họa, giúp **lập bản đồ vùng nguy hiểm**, **giám sát thời gian thực**, và **lập kế hoạch sơ tán**,. Việc tích hợp dữ liệu địa không gian thời gian thực vào các mô hình AI dẫn đến **dự đoán và đánh giá tốt hơn**.

#### III. Xác định những tồn tại và hạn chế của công bố liên quan

Mặc dù các mô hình lai và tích hợp đã đạt được độ chính xác cao, các nguồn vẫn xác định những hạn chế quan trọng cần được giải quyết trong nghiên cứu tương lai:

| Hạn chế/Tồn tại | Chi tiết Hạn chế | Nguồn |
| :--- | :--- | :--- |
| **Vấn đề Dữ liệu Thiểu số và Mất cân bằng lớp (Class Imbalance)** | Các mô hình, dù có sử dụng SMOTE, vẫn gặp khó khăn khi dự đoán các **thảm họa hiếm gặp hoặc cục bộ** (rare or localized disasters, ví dụ: vi trượt lở). Mô hình có xu hướng bị **thiên vị (biased)** đối với các lớp đa số (ví dụ: lũ lụt),,. Đối với TCI, các bão cực mạnh (Cấp 6/SuperTY) hiếm gặp dẫn đến **thiếu dữ liệu đào tạo** và sai sót trong phân loại cường độ. |,,,, |
| **Chất lượng và Tính toàn vẹn của Dữ liệu (Data Quality)** | Hiệu suất mô hình bị ảnh hưởng khi dữ liệu địa không gian **thưa thớt, nhiễu** (sparse or noisy geospatial data), hoặc **ảnh vệ tinh lỗi thời** (outdated satellite imagery),. Các cơ sở dữ liệu toàn cầu (như EM-DAT) thường **thiếu dữ liệu nghiêm trọng** về kinh tế, thiệt hại hoặc tọa độ, dẫn đến ước tính thiên vị,. |,,, |
| **Tính Tổng quát hóa (Generalizability)** | Việc đào tạo mô hình chỉ trên dữ liệu của một khu vực cụ thể (ví dụ: Tây Bắc Thái Bình Dương cho TCI, hoặc Bồn địa sông Danshui) có thể **hạn chế tính tổng quát hóa** của mô hình khi áp dụng cho các vùng địa lý hoặc khí hậu khác,. |, |
| **Yêu cầu về Tính toán (Computational Demands)** | Các kiến trúc Học sâu phức tạp (ví dụ: TCIque) đòi hỏi **nguồn lực tính toán đáng kể** (substantial computational resources), làm tăng chi phí vận hành và hạn chế ứng dụng trong môi trường có khả năng hạn chế,,. |,, |
| **Hạn chế về Dữ liệu Thời gian thực (Real-time Data)** | Việc áp dụng thực tế bị thách thức do **khó khăn trong việc thu thập dữ liệu đa phương thức cần thiết theo thời gian thực** và xử lý dữ liệu bị thiếu hoặc bị trì hoãn. Ví dụ, mô hình N-XGB hiện đang tối ưu hóa cho tập dữ liệu dạng bảng và **chưa xem xét các nguồn thời gian thực** như hình ảnh vệ tinh hoặc dữ liệu cảm biến IoT. |,, |
